```python
from pylab import *

from itertools import product
import numpy as np
np.random.seed(0)

import nengo
from nengo.utils.distributions import Uniform, UniformHypersphere

n_x = 200
n_y = 200
n_m = 1000
n_dope = 50
d_x = 4
d_y = 4
d_m = d_x
intercept_spread = 0.05
neurons_per_item = 20
assert d_x <= d_m
assert d_y <= d_m

def keys(t):
    r = 2.0
    if t < r*1:
        return [1.0, 0, 0, 0]
    if t < r*2:
        return [0, 1.0, 0, 0]
    if t < r*3:
        return [0.0, -1.0, 0, 0]
    if t < r*4:
        return [0, 0, 0, -1.0]
    def k(t):
        if t < r*5:
            return [1.0, 0, 0, 0]
        if t < r*6:
            return [0, 1.0, 0, 0]
        if t < r*7:
            return [0.0, -1.0, 0, 0]
        if t < r*8:
            return [0, 0, 0, -1.0]
    return np.asarray(k(t)) + np.random.normal(0, 0.1, 4)

def values(t):
    r = 2.0
    if t < r*1:
        return [0.0, -1.0, 0, 0]
    if t < r*2:
        return [0.7, 0, -0.7, 0]
    if t < r*3:
        return [1.0, 0, 0, 0]
    if t < r*4:
        return [0.0, 0.5, 0, 0.5]
    return [0, 0, 0, 0]

def is_learning(t):
    return [int(t < 8)]

def calculate_intercept(d, cutoff, dx = 0.001):
    x = np.arange(-1+dx, 1, dx)
    y = (1 - x**2)**((d - 3)/2.0)
    py = np.cumsum(y)
    py = py / sum(py) / dx
    return x[py >= 1 - cutoff][0] 

intercept = calculate_intercept(d_m, float(neurons_per_item)/n_m)
mem_intercepts = Uniform(intercept - intercept_spread, min(intercept + intercept_spread, 1.0))
learn_intercepts = Uniform(0.1, 1.0)

print intercept

m = nengo.Network()
with m:
    key = nengo.Node(output=keys)
    value = nengo.Node(output=values)
    learning = nengo.Node(output=is_learning)
    bias = nengo.Node(output=[1])
    dopamine = nengo.Ensemble(nengo.LIF(n_dope), 1, intercepts=learn_intercepts, encoders=[[1]]*n_dope)
    nopamine = nengo.Ensemble(nengo.LIF(n_dope), 1, intercepts=learn_intercepts, encoders=[[1]]*n_dope)
    x = nengo.Ensemble(nengo.LIF(n_x), d_x, label="x")
    mem = nengo.Ensemble(nengo.LIF(n_m), d_m, label="mem", intercepts=mem_intercepts)
    mem_static = nengo.Ensemble(nengo.LIF(n_m), d_m, label="mem", intercepts=mem_intercepts)
    y_est = nengo.Ensemble(nengo.LIF(n_y), d_y, label="y_est")
    e = nengo.Ensemble(nengo.LIF(n_y), d_y, label="error")

def inhibit(pre, post, amount=20, filter=0.001, **kwargs):
    return nengo.Connection(
        pre.neurons, post.neurons,
        transform=-amount*np.ones((post.n_neurons, pre.n_neurons)), 
        **kwargs)

def spread_sample(sampler, n, thresh):
    p = []
    for _ in range(n):
        for attempt in range(100):
            v = sampler.sample(1)[0]
            if not True in (np.linalg.norm(u - v) < thresh for u in p):
                break
        else:
            raise RuntimeError('thresh=%s too high' % thresh)
        p.append(v)
    return p

def autoconnect(pre, post, **kwargs):
    eval_points = spread_sample(UniformHypersphere(pre.dimensions, surface=True), post.dimensions, 0.4)
    eval_points_lookup = dict(((tuple(v), i) for i, v in enumerate(eval_points)))
    function = lambda v: np.eye(post.dimensions)[eval_points_lookup[tuple(v)], :]
    return nengo.Connection(pre, post, eval_points=eval_points, function=function, **kwargs)

with m:
    nengo.Connection(key, x)
    nengo.Connection(x, mem, learning_rule=nengo.OJA(learning_rate=5e-3, learning=dopamine, scale=2.0, post_tau=0.001))
    nengo.Connection(x, mem_static)
    nengo.Connection(value, e)
    nengo.Connection(y_est, e, transform=-1)
    nengo.Connection(learning, dopamine)
    nengo.Connection(bias, nopamine)
    inhibit(dopamine, nopamine)
    inhibit(nopamine, e)
    autoconnect(mem, y_est, learning_rule=nengo.PES(e, learning_rate=1e-3))

    p_key = nengo.Probe(key, "output", filter=0.03)
    p_value = nengo.Probe(value, "output", filter=0.03)
    p_x = nengo.Probe(x, "decoded_output", filter=0.03)
    spikes_x = nengo.Probe(x, "spikes")
    p_mem_static = nengo.Probe(mem_static, "decoded_output", filter=0.03)
    spikes_mem_static = nengo.Probe(mem_static, "spikes")
    p_mem = nengo.Probe(mem, "decoded_output", filter=0.03)
    spikes_mem = nengo.Probe(mem, "spikes")
    p_e = nengo.Probe(e, "decoded_output", filter=0.03)
    p_dope = nengo.Probe(dopamine, "decoded_output", filter=0.03)
    p_nope = nengo.Probe(nopamine, "decoded_output", filter=0.03)
    p_y_est = nengo.Probe(y_est, "decoded_output", filter=0.03)
    #spikes_y_est = nengo.Probe(y_est, "spikes")
```

```python
from nengo.utils.matplotlib import rasterplot

sim = nengo.Simulator(m)
sim.run(16)
```

```python
figure()
title('Key')
plot(sim.data[p_key])

figure()
title('Value')
plot(sim.data[p_value])

figure()
title('x Spikes')
rasterplot(sim.trange(), sim.data[spikes_x])

figure()
title('Static Memory')
plot(sim.data[p_mem_static])

figure()
title('Static Memory Spikes')
rasterplot(sim.trange(), sim.data[spikes_mem_static])

figure()
title('Memory')
plot(sim.data[p_mem])

figure(figsize=(6,12))
title('Memory Spikes')
rasterplot(sim.trange(), sim.data[spikes_mem])

figure()
title('y')
plot(sim.data[p_y_est])

#figure()
#title('y Spikes')
#rasterplot(sim.trange(), sim.data(spikes_y_est))

figure()
title('Error')
plot(sim.data[p_e], label='Error')
plot(sim.data[p_dope], label='Dopamine')
plot(sim.data[p_nope], label='No Learning')
legend(loc='best')
show()
```
