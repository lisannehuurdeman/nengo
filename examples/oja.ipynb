```python
from itertools import product
import numpy as np

import nengo

n_x = 200
n_y = 200
n_m = 1000
n_dope = 50
d_x = 4
d_y = 4
d_m = max(d_x, 4)
intercepts = nengo.objects.Uniform(0.1, 1.0)

def keys(t):
    if t < 1:
        return [1.0, 0, 0, 0]
    if t < 2:
        return [0, 1.0, 0, 0]
    if t < 3:
        return [0.0, -1.0, 0, 0]
    if t < 4:
        return [0.7, 0, -0.7, 0]
    def k(t):
        if t < 5:
            return [1.0, 0, 0, 0]
        if t < 6:
            return [0, 1.0, 0, 0]
        if t < 7:
            return [0.0, -1.0, 0, 0]
        if t < 8:
            return [0.7, 0, -0.7, 0]
    return np.asarray(k(t)) + np.random.normal(0, 0.1, 4)

def values(t):
    if t < 1:
        return [0.0, -1.0, 0, 0]
    if t < 2:
        return [0.7, 0, -0.7, 0]
    if t < 3:
        return [1.0, 0, 0, 0]
    if t < 4:
        return [0.0, 0.5, 0, 0.5]
    return [0, 0, 0, 0]

def is_learning(t):
    return [int(t < 4)]

m = nengo.Model()
key = nengo.Node(output=keys)
value = nengo.Node(output=values)
learning = nengo.Node(output=is_learning)
bias = nengo.Node(output=[1])
dopamine = nengo.Ensemble(nengo.LIF(n_dope), 1, intercepts=intercepts, encoders=[[1]]*n_dope)
nopamine = nengo.Ensemble(nengo.LIF(n_dope), 1, intercepts=intercepts, encoders=[[1]]*n_dope)
x = nengo.Ensemble(nengo.LIF(n_x), d_x, label="x")
y = nengo.Ensemble(nengo.LIF(n_m), d_m, label="y", intercepts=intercepts)
y_est = nengo.Ensemble(nengo.LIF(n_y), d_y, label="y_est")
e = nengo.Ensemble(nengo.LIF(n_y), d_y, label="error")

def inhibit(pre, post, amount=20, filter=0.001):
    nengo.Connection(pre.neurons, post.neurons, transform=-amount*np.ones((post.n_neurons, pre.n_neurons)), filter=filter)

def copy(x, d):
    """Extends x with 'd - len(x)' copies."""
    return np.asarray(list(x)*(d/len(x) + 1))[:d]

def polynomial_features(x, d):
    """Extends x with 'd - len(x)' multivariate-polynomials in x of increasing degree."""
    def go(x, d):
        fx = []
        degree = 1
        while True:
            for elems in product(range(len(x)), repeat=degree+1):
                p = 1
                for elem in elems:
                    p *= x[elem]
                fx.append(p)
                if len(fx) >= d:
                    return fx
            degree += 1
    fx = np.asarray(go(x, d))
    return fx / np.dot(fx, fx)**0.5  # normalize


assert d_x <= d_m
assert d_y <= d_m
nengo.Connection(key, x)
nengo.Connection(x, y, transform=0.5, # function=lambda v: polynomial_features(v, d=d_m), 
    learning_rule=nengo.OJA(learning_rate=1e-1, learning=dopamine, scale=1.0))
nengo.Connection(value, e)
nengo.Connection(y_est, e, transform=-1)
nengo.Connection(learning, dopamine)
nengo.Connection(bias, nopamine)
inhibit(dopamine, nopamine)
inhibit(nopamine, e)
nengo.Connection(y, y_est, function=lambda v: v[:d_y],
    learning_rule=nengo.PES(e, learning_rate=1e-3))

p_key = nengo.Probe(key, "output", filter=0.03)
p_value = nengo.Probe(value, "output", filter=0.03)
p_x = nengo.Probe(x, "decoded_output", filter=0.03)
spikes_x = nengo.Probe(x, "spikes")
p_y = nengo.Probe(y, "decoded_output", filter=0.03)
spikes_y = nengo.Probe(y, "spikes")
p_e = nengo.Probe(e, "decoded_output", filter=0.03)
p_dope = nengo.Probe(dopamine, "decoded_output", filter=0.03)
p_nope = nengo.Probe(nopamine, "decoded_output", filter=0.03)
p_y_est = nengo.Probe(y_est, "decoded_output", filter=0.03)
#spikes_y_est = nengo.Probe(y_est, "spikes")
```

```python
from nengo.matplotlib import rasterplot

sim = nengo.Simulator(m)
sim.run(4)
```

```python
figure()
title('Key')
plot(sim.data[p_key])

figure()
title('Value')
plot(sim.data[p_value])

figure()
title('x Spikes')
rasterplot(sim.trange(), sim.data[spikes_x])

figure()
title('Memory')
plot(sim.data[p_y])

figure()
title('Memory Spikes')
rasterplot(sim.trange(), sim.data[spikes_y])

figure()
title('y')
plot(sim.data[p_y_est])

#figure()
#title('y Spikes')
#rasterplot(sim.trange(), sim.data(spikes_y_est))

figure()
title('Error')
plot(sim.data[p_e], label='Error')
plot(sim.data[p_dope], label='Dopamine')
plot(sim.data[p_nope], label='No Learning')
legend(loc='best')
show()
```
